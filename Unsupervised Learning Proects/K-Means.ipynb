```python
X = pd.get_dummies(df).astype(int)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
# help(KMeans)
```

```python
model = KMeans(n_clusters=2)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels
```

```python
X['Cluster'] = cluster_labels
```

```python
X
```

```python
X.corr()['Cluster'].iloc[:-1].sort_values()
```

```python
plt.figure(figsize=(12,6),dpi=200)
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar')
```

## Using the eldow method to know which k value is better.

```python
ssd = []

for k in range(2,10):
    
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    
    
    ssd.append(model.inertia_) # Sum of Square Distances(SSD) point --> cluster center
```

```python
ssd
```

```python
plt.plot(range(2,10),ssd,'o--')
```

## to calculate the differences between one row and the other row using pandas.

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff()
```

# KMeans Color Quantization

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_as_array = mpimg.imread('./DATA/palm_trees.jpg')
```

```python
image_as_array
```

```python
image_as_array.shape
```

```python
plt.figure(dpi=120)
plt.imshow(image_as_array)
```

```python
# converting image to 2D
(h,w,c) = image_as_array.shape
```

```python
image_as_array2d = image_as_array.reshape(h*w,c)
```

```python
image_as_array
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6)
```

```python
labels = model.fit_predict(image_as_array2d)
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
quantized_image = np.reshape(rgb_codes[labels],(h,w,c))
```

```python
quantized_image
```

```python
plt.imshow(quantized_image)
```
